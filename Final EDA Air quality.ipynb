```python
# Import Pandas library
import pandas as pd

# Reading the csv file and passing it to dataframe
df = pd.read_csv("cities_air_quality_water_pollution.18-10-2021 (1).csv")

# viewing the first five rows
df.head()
```

```python
# Checking for null values
df.isna()
```

```python
# checking null values for each column
df.isna().sum()
```

```python
# checking the columns
df.columns
```

```python
# Removing double quotes from all column names
df.columns = df.columns.str.replace('"', '')
df
```

```python
# Removing double quotes from Region columns
df[' Region'] = df[' Region'].str.replace('"', '')
```

```python
# Removing double quotes from Country columns
df[' Country'] = df[' Country'].str.replace('"', '')
```

```python
df
```

```python
# Checking if there are missing values in the dataframe
df.isnull().values.any()
```

```python
# Checking for missing or empty values in the entire dataframe
df.isnull()
```

```python
# Check for empty strings in the DataFrame
(df == '').sum()
```

```python
# Check for missing values or strings with only spaces
df.apply(lambda x: x.str.strip() == '' if x.dtype == 'object' else False).sum()
```

```python
# Checking for duplicates
df.duplicated()
```

```python
# Checking for duplicates
df.duplicated().sum()
```

```python
# Checking the data types
df.dtypes
```

```python
# Checking for quick statistical summary
df.describe()
```

```python
# Replace missing values in 'Region' with 'Unknown'
df[' Region'].fillna('Unknown', inplace=True)

print(df)
```

```python
df
```

```python
# Replace empty strings, strings with only spaces, and NaN with 'Unknown'
df[' Region'].replace(r'^\s*$', 'Unknown', regex=True, inplace=True)
df
```

```python
import matplotlib.pyplot as plt

# Create a boxplot for 'AirQuality'
plt.boxplot(df[' AirQuality'].dropna())  # Drop missing values for better plot
plt.title('AirQuality Boxplot')
plt.show()
```

```python
# Create a boxplot for 'AirQuality'
plt.boxplot(df[' WaterPollution'].dropna())  # Drop missing values for better plot
plt.title('WaterPollution Boxplot')
plt.show()
```

```python
# Scatter plot for 'AirQuality' vs 'WaterPollution'
df.plot(kind='scatter', x=' AirQuality', y=' WaterPollution', color = "orange", title='AirQuality vs WaterPollution')
plt.show()
```

```python
Low_air_quality = df.groupby(' Country')[' AirQuality'].mean().sort_values(ascending=True).head(20)

plt.figure(figsize=(10,5))

sns.barplot(x=Low_air_quality.index, y=Low_air_quality)
plt.title("Countries With Low Air Quality")
plt.xlabel("Country")
plt.ylabel("Air Quality")

plt.xticks(rotation=90)
plt.show()
```

```python
#lets visualize a bar chat for top 20 air polutants
bottom_20_air = df.groupby(' Country')[' AirQuality'].mean().sort_values(ascending=True).head(20)

#lets plot a barchart
#lets define figure size
plt.figure(figsize=(10,6))

#lets plot our graph using seaborn
sns.barplot(x=bottom_20_air.index, y=bottom_20_air )
plt.xticks(rotation=90)
plt.title("Top 20 countries with dirty air")
plt.show()
```

```python
High_air_quality = df.groupby(' Country')[' AirQuality'].mean().sort_values(ascending=False).head(20)

plt.figure(figsize=(10,5))

sns.barplot(x=High_air_quality.index, y=High_air_quality)
plt.title("Countries With High Air Quality")
plt.xlabel("Country")
plt.ylabel("Air Quality")

plt.xticks(rotation=90)
plt.show()
```

```python
Low_Water_Pollution = df.groupby(' Country')[' WaterPollution'].mean().sort_values(ascending=True).head(20)

plt.figure(figsize=(10,5))

sns.barplot(x=Low_Water_Pollution.index, y=Low_Water_Pollution)
plt.title("Countries With Low Water Pollutiion")
plt.xlabel("Country")
plt.ylabel("Water Pollution")

plt.xticks(rotation=90)
plt.show()
```

```python
High_Water_Pollution = df.groupby(' Country')[' WaterPollution'].mean().sort_values(ascending=False).head(20)

plt.figure(figsize=(10,5))

sns.barplot(x=High_Water_Pollution.index, y=High_Water_Pollution)
plt.title("Countries With High Water Pollutiion")
plt.xlabel("Country")
plt.ylabel("Water Pollution")

plt.xticks(rotation=90)
plt.show()
```

```python
#lets plot a heatmap of water and air
water_air =[' AirQuality',' WaterPollution']
df1=df[water_air].corr()
# Plotting a heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(df1, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap of the AirQuality and WaterPollution')
plt.show()
```
